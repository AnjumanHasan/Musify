```python
import pandas as pd
import numpy as np
import Recommenders as Recommenders
import pickle
import json
```

```python
song_df_1 = pd.read_csv('triplets_file.csv')
song_df_1.head()
```

```python
song_df_2 = pd.read_csv('song_data.csv')
song_df_2.head()
```

```python
##Preprocessing the Data
```

```python
song_df = pd.merge(song_df_1, song_df_2.drop_duplicates(['song_id']), on='song_id', how='left')
song_df.head()
```

```python
song_df['song'] =song_df['title'] + "-" + song_df['artist_name']
song_df.head()
```

```python
song_df = song_df.head(10000)
```

```python
song_grouped = song_df.groupby(['song']).agg({'listen_count':'count'}).reset_index()
song_grouped.head()
```

```python
grouped_sum = song_grouped['listen_count'].sum()
song_grouped['percentage'] = (song_grouped['listen_count'] / grouped_sum ) * 100
song_grouped.sort_values(['listen_count', 'song'], ascending=[0,1])
```

```python
##Recommendations : Popularity Based
```

```python
pr = Recommenders.popularity_recommender_py()
```

```python
pr.create(song_df, 'user_id', 'song')
```

```python
# displaying top 10 popular songs
#high listen count numbers
pr.recommend(song_df['user_id'][5])
```

```python
pr.recommend(song_df['user_id'][100])
```

```python
#same songs are recommended as popularity of songs do not depend on individuals
```

```python
##Item Based recommendations.
#Content Based filtering
```

```python
ir = Recommenders.item_similarity_recommender_py()
ir.create(song_df, 'user_id', 'song')
```

```python
user_items = ir.get_user_items(song_df['user_id'][5])
```

```python
for user_item in user_items:
    print(user_item)
```

```python
ir.recommend(song_df['user_id'][5])
```

```python
#the model has recommended the most accurate songs for a particular user
#testing for other users
user_items = ir.get_user_items(song_df['user_id'][100])
```

```python
for user_item in user_items:
    print(user_item)
```

```python
ir.recommend(song_df['user_id'][100])
```

```python
# give related songs based on the words
ir.get_similar_items(['Love Story-Taylor Swift'])
```

```python
with open ('recommender.pickle','wb') as f :
    pickle.dump(ir,f)
```
